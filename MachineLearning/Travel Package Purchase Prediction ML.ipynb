```python
cat1 = df[(df['Age'] >= 18) & (df['Age'] <=30)]
cat2 = df[(df['Age'] > 30) & (df['Age'] <=50)]
cat3 = df[(df['Age'] > 50)]
combined_cat = pd.concat([cat1, cat2, cat3], keys=['18-30', '31-50', '51+'])
```

```python
plt.figure(figsize=(20,15))
sns.countplot(x='Age',hue='ProdTaken',data=combined_cat)
```

1) From the above chart, we can see that people between the ages of 30-40 aren't satisfied with the travel package and they haven't taken the product.

2) People between the ages of 18-21 have taken the product more number of times than rejecting it. So the travel package is quite popular among young people.

3) Overall chances of product being taken is very less i.e. most of the people aren't taking the product. So you need to focus on making a better travel package.

4) Final conclusion is that, the plan has to be changed according to the middle aged people who are mostly ageing between 30-40.

Feature Selection

```python
# Correlation Threshold
cols = list(df.columns)
selected_features = []
for col in cols:
  if col != 'ProdTaken':
    if ((np.abs(df['ProdTaken'].corr(df[col])) > 0.05)):
      selected_features.append(col)
print(selected_features)
print(len(selected_features))
```

```python
# Lasso Technique
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
Las = Lasso(alpha=0.1)
X = df.drop(columns=['ProdTaken'])
y = df['ProdTaken']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
Las.fit(X_train,y_train)
feature_coefficients = pd.Series(Las.coef_, index=X.columns)
selected_las_features = feature_coefficients[feature_coefficients != 0].index.tolist()
print(selected_las_features)
```

Model Building


Logistic Regression

```python
df_scaled = df.drop(columns='ProdTaken')
df_scaled = pd.DataFrame(sc.fit_transform(df_scaled),columns=df_scaled.columns)
df_scaled.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
Logreg = LogisticRegression()
X = df_scaled
y = df['ProdTaken']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
model = Logreg.fit(X_train,y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
print(f'Train Accuracy: {accuracy_score(y_train,y_train_pred)}')
print(f'Test Accuracy: {accuracy_score(y_test,y_test_pred)}')
```

```python
cv_scores = cross_val_score(model, X, y, cv=5)
for i in cv_scores:
  print(i)
```

```python
average_cv_score = cv_scores.mean()
print(f"Average cross-validation score: {average_cv_score:.2f}")
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
X = df.drop(columns='ProdTaken')
y = df['ProdTaken']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
c_parameter_name = 'Max_Depth'
c_parameter_values = list(range(1,50))
df_acc_test = pd.DataFrame(columns=[c_parameter_name])
for i in c_parameter_values:
  model = DecisionTreeClassifier(max_depth=i,splitter='best')
  y_pred_train = model.fit(X_train,y_train)
  y_pred = model.predict(X_test)
  acc_score = accuracy_score(y_test,y_pred)*100
  df_acc_test = df_acc_test.append({c_parameter_name : i, 'Accuracy' : acc_score},ignore_index=True)
df_acc_test
```

```python
def report(y1,y2,dom):
  print(f'{dom} Confusion Matrix')
  cm = confusion_matrix(y1,y2)
  sns.heatmap(cm,annot=True,fmt='g')
  plt.tight_layout()
  plt.show()
  print(accuracy_score(y1,y2))
```

```python
dtc = DecisionTreeClassifier(max_depth=15,splitter='best')
model = dtc.fit(X_train,y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
report(y_train,y_train_pred,dom='Training')
```

```python
report(y_test,y_test_pred,dom='Testing')
```

Random Forest Classifier

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
X = df.drop(columns='ProdTaken')
y = df['ProdTaken']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
n_estimators = [int(i) for i in np.linspace(start=200,stop=2000,num=10)]
max_features = ['auto','sqrt']
max_depth = [int(i) for i in np.linspace(10,110,num=11)]
min_samples_split = [2,5,10]
min_samples_leaf = [1,2,4]
bootstrap = [True,False]
random_grid = {'n_estimators' : n_estimators, 'max_depth' : max_depth, 'min_samples_split' : min_samples_split, 'min_samples_leaf' : min_samples_leaf, 'bootstrap' : bootstrap}
print(random_grid)
```

```python
rfc = RandomForestClassifier(random_state = 23)
rfc_random = RandomizedSearchCV(estimator=rfc,param_distributions=random_grid,n_jobs=-1,scoring='neg_mean_absolute_error',random_state=23,cv=3, return_train_score = True)
rfc2 = rfc_random.fit(X_train,y_train)
```

```python
rfc2.best_params_
```

```python
y_train_pred = rfc2.predict(X_train)
y_test_pred = rfc2.predict(X_test)
```

```python
report(y_train,y_train_pred,dom='Training')
```

```python
report(y_test,y_test_pred,dom='Testing')
```

KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
X = df_scaled
y = df['ProdTaken']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,marker='o')
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=22)
knn_model = knn.fit(X_train,y_train)
y_train_pred = knn_model.predict(X_train)
y_test_pred = knn_model.predict(X_test)
```

```python
report(y_train,y_train_pred,dom='Training')
```

```python
report(y_test,y_test_pred,dom='Testing')
```
